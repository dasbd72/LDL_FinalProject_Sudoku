```python
import numpy as np
SIZE = 28
VALIDATION_SPLIT = 0.2
BATCH_SIZE = 32
EPOCHS = 1000
```

```python
from sklearn.utils import shuffle
import pandas as pd
from tensorflow import keras

df = pd.read_csv('../Dataset/dataset.csv')
df = shuffle(df)
mid = int(df.shape[0] * (1 - VALIDATION_SPLIT))
x_train = np.array(df.iloc[:mid,1:])
x_valid = np.array(df.iloc[mid:,1:])
y_train = np.array(df.iloc[:mid,0])
y_valid = np.array(df.iloc[mid:,0])

x_train = x_train.reshape(-1, SIZE, SIZE, 1)
x_valid = x_valid.reshape(-1, SIZE, SIZE, 1)
num_categories = 10
y_train = keras.utils.to_categorical(y_train, num_categories)
y_valid = keras.utils.to_categorical(y_valid, num_categories)

print(x_train.shape)
print(x_valid.shape)
print(y_train.shape)
print(y_valid.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
model = Sequential()
model.add(Flatten(input_shape=(SIZE, SIZE, 1)))
model.add(Dense(units = 64, activation='relu'))
model.add(Dense(units = num_categories, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=10,  # randomly rotate images in the range (degrees, 0 to 180)
    zoom_range=0.1,  # Randomly zoom image
    width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
    height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
    validation_split=0.2
)
```

```python

datagen.fit(x_train)
img_iter = datagen.flow(x_train, y_train, batch_size=BATCH_SIZE)
```

```python
history = model.fit(
    img_iter,
    validation_data = (x_valid, y_valid),
    epochs = EPOCHS
)
```

```python
model.save('Number_Model')
```
